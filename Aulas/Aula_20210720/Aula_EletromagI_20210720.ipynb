## (Revisão e análise) Exemplo 1 : Anel circular carregado eletricamente

### 1.a - Calcule o campo elétrico $\vec{E}$ criado por um anel circular carregado eletricamente;

Aproveitando a simetria do anel circular e colocando-o no plano $xy$, centrado na origem $O$ :

![image-2.png](attachment:image-2.png "Anel circular com cargas elétricas")

Para calcular o campo elétrico $\vec{E}$ criado por um anel circular de raio $R$ carregado eletricamente, usamos a expressão com densidade linear de carga elétrica $\lambda$ $(\propto C/m)$ dada em uma curva $C$ :

$$ \vec{E}(\vec{r}_0) = \frac{1}{4\pi\epsilon_0}\int_C \frac{\lambda(\vec{r}')(\vec{r}_0-\vec{r}')\,ds'}{|\vec{r}_0-\vec{r}'|^3} $$

Assim, em coordenadas cilíndricas, ao longo da circunferência de raio $R$ centrada na origem :

 

$$\textrm{curva}\,C : \begin{cases}
\rho' = R \\\\
\theta' \in [0, 2\,\pi]\,\textrm{rad} \\\\
z' = 0\,\textrm{m}
\end{cases}$$

Logo temos que $\vec{E}$ é uma integral em relação a $\theta'$ :

$$ \rightarrow \,\,\, \vec{E}(\vec{r}_0) = \frac{1}{4\pi\epsilon_0}\int_0^{2\,\pi} \frac{\lambda(\theta')(\vec{r}_0-\vec{r}')\,R\,d\theta'}{|\vec{r}_0-\vec{r}'|^3} $$

$$ \rightarrow \,\,\, \vec{E}(\vec{r}_0) = \frac{R}{4\pi\epsilon_0}\int_0^{2\,\pi} \frac{\lambda(\theta')[(x_0 - R\cos\theta')\hat{i}
+ (y_0 - R\sin\theta')\,\hat{j} + z_0\,\hat{k}]\,d\theta'}{[r_0^2 + R^2 - 2R(x_0\cos\theta' + y_0\sin\theta')]^{3/2}} $$

Expandindo em 3 integrais simples escalares, uma para cada componente tridimensional de $\vec{E} = \vec{E}_x\hat{i} + \vec{E}_y\hat{j} + \vec{E}_z\hat{k}$ :

$$ E_x(\vec{r}_0) = \frac{R}{4\pi\epsilon_0}\int_0^{2\,\pi} \frac{\lambda(\theta')(x_0 - R\cos\theta')
\,d\theta'}{[r_0^2 + R^2 - 2R(x_0\cos\theta' + y_0\sin\theta')]^{3/2}} $$

$$ E_y(\vec{r}_0) = \frac{R}{4\pi\epsilon_0}\int_0^{2\,\pi} \frac{\lambda(\theta')(y_0 - R\sin\theta')\,d\theta'}{[r_0^2 + R^2 - 2R(x_0\cos\theta' + y_0\sin\theta')]^{3/2}} $$

$$ E_z(\vec{r}_0) = \frac{R}{4\pi\epsilon_0}\int_0^{2\,\pi} \frac{\lambda(\theta')z_0\,\hat{k}\,d\theta'}{[r_0^2 + R^2 - 2R(x_0\cos\theta' + y_0\sin\theta')]^{3/2}} $$

A solução acima será abordada depois ao praticarmos gráficos de campos vetoriais 3D.

### 1.b - Calcule $\vec{E}$ criado por um anel circular carregado eletricamente, com observador ao longo do eixo do anel

Ao restringir o observador (posição da carga-teste) ao longo do eixo $z$ do anel (eixo ortogonal ao plano do anel circular), temos :

$$ \vec{E}(\vec{r}_0) = \vec{E}(z_0) = \frac{R}{4\pi\epsilon_0 (z_0^2 + R^2)^{3/2}}\int_0^{2\,\pi} \lambda(\theta')[(- R\cos\theta')\hat{i}
+ (- R\sin\theta')\,\hat{j} + z_0\,\hat{k}]\,d\theta' $$

que é uma integral vetorial 3D (igual a 3 integrais simples).

A solução acima, escolhendo certa função não constante para $\lambda(\theta')$, será abordada depois ao praticarmos gráficos de campos vetoriais 3D.

### 1.c - Calcule $\vec{E}$ criado por um anel circular carregado eletricamente de forma homogênea

Ao restringir para distribuição homogênea de carga elétrica no anel, i. e., $\lambda(\vec{r}') = \lambda_0 = \textrm{cte}$, temos uma leve simplificação da integral de (1.a) :

$$ \vec{E}(\vec{r}_0) = \frac{\lambda_0 R}{4\pi\epsilon_0}\int_0^{2\,\pi} \frac{[(x_0 - R\cos\theta')\hat{i}
+ (y_0 - R\sin\theta')\,\hat{j} + z_0\,\hat{k}]\,d\theta'}{[r_0^2 + R^2 - 2R(x_0\cos\theta' + y_0\sin\theta')]^{3/2}} $$

A solução acima será abordada depois ao praticarmos gráficos de campos vetoriais 3D.

### 1.d -  Calcule $\vec{E}$ criado por um anel circular carregado eletricamente de forma homogênea, com observador ao longo do eixo do anel

Ao restringir para distribuição homogênea de carga elétrica no anel, i. e., $\lambda(\vec{r}') = \lambda_0 = \textrm{cte}$, bem como tendo o observador (posição da carga-teste) ao longo do eixo $z$ do anel (eixo ortogonal ao plano do anel circular), temos que $\vec{E}$ só tem uma componente vetorial, ao longo do eixo $z$, as outras componentes vetoriais são nulas devido à simetria do problema :

$$ \vec{E}(z_0) = \frac{\lambda_0 R\,z_0}{2\epsilon_0 (z_0^2 + R^2)^{3/2}} \hat{k} $$

Ou, em termos da carga total $Q = 2\pi R \lambda_0 $ :

$$ \vec{E}(z_0) = \frac{Q\,z_0}{4\pi\epsilon_0 (z_0^2 + R^2)^{3/2}} \hat{k} $$

#### Análise da solução (1.d) de $\vec{E}(z_0)$

A função é impar em relação a $z_0$, logo, para carga $Q$ positiva, $E(z_0)$ tem o mesmo sinal que $z_0$.

Para observador (carga-teste) no plano e centro do anel, $z_0 = 0$ m, temos que o campo elétrico se anula devido à simetria de cargas elétricas em volta do observador :

$$ \vec{E}(z_0 = 0 \text{ m}) = \left( 0\,\frac{N}{C} \right) \,\hat{k} $$

Para observador (carga-teste) longe da origem, $z_0 >> R$, vamos calcular o limite quando $R/z_0$ tende a valor infinitesimal :

$$ \lim_{R/z_0 \to 0} \vec{E}(z_0) = \lim_{R/z_0 \to 0} \frac{Q\,z_0}{4\pi\epsilon_0 (z_0^2 + R^2)^{3/2}} \hat{k} = \lim_{R/z_0 \to 0} \frac{Q\,z_0}{4\pi\epsilon_0 |z_0|^3(1 + R^2/z_0^2)^{3/2}} \hat{k} = \frac{Q\,z_0}{4\pi\epsilon_0 |z_0|^3} \hat{k}$$ 

Ora, esse é o campo elétrico de uma carga elétrica pontual $Q$ na origem, medido ao longo do eixo $z$ na posição $\vec{r}_0 = z_0\,\hat{k}$ :

$$ \vec{E}(\vec{r}_0) = \vec{E}(z_0) = \frac{Q\,\vec{r}_0}{4\pi\epsilon_0\,r_0^3} = \frac{Q\,z_0}{4\pi\epsilon_0\,|z_0|^3}\hat{k} $$

Ou seja, o anel circular carregado eletricamente de forma homogênea, com observador ao longo do eixo do anel e muito longe do anel, mede um campo elétrico do anel semelhante a se tivesse toda a carga elétrica $Q$ no centro do anel.

## (Continuação) Exemplo 2 : Disco circular carregado eletricamente

O campo elétrico $\vec{E}$ para o caso genérico de disco circular carregado eletricamente pode ser resolvido via integral de superfície com densidade superficial de carga elétrica, $\sigma(\vec{r}')$ na superfície $S$ do disco :

$$ \vec{E}(\vec{r}_0) = \frac{1}{4\pi\epsilon_0}\int\int_S \frac{\sigma(\vec{r}')\,\vec{r}\,dA'}{|\vec{r}|^3} = \frac{1}{4\pi\epsilon_0}\int\int_S \frac{\sigma(\vec{r}')(\vec{r}_0-\vec{r}')\,dA'}{|\vec{r}_0-\vec{r}'|^3} $$

Pode-se usar as mesmas restrições do exemplo (1), anel circular carregado eletricamente, como :

- o observador (posição da carga-teste) ao longo do eixo $z$ do anel (eixo ortogonal ao plano do disco circular);

- distribuição homogênea de carga elétrica no disco, i. e., densidade superficial de carga elétrica, $\sigma(\vec{r}') = \sigma_0 = \text{cte} $; 

logo teríamos analogamente casos (2.a), (2.b), (2.c) e (2.d). Vamos focar no caso (2.d).

### 2.d -  Calcule $\vec{E}$ criado por um disco circular carregado eletricamente de forma homogênea, com observador ao longo do eixo do anel

![image.png](attachment:image.png "Disco circular carregado eletricamente")

#### Usando duas integrais simples no lugar de uma integral dupla

Ao invés de calcular a intgral dupla de superfície, pode-se trocar por duas integrais simples, sendo a primeira feita no cálculo anterior (1.d) de anel circular carregado eletricamente de forma homogênea. A 2a integral usa o resultado do anel circular, substituindo $R$ por $\rho'$ e calculando o diferencial em relação a $Q$ para se ter o diferencial de campo elétrico :

$$ d\vec{E} = \frac{dQ\,z_0}{4\pi\epsilon_0 (z_0^2 + \rho'^2)^{3/2}} \hat{k} $$

onde no disco $dQ = \sigma_0\,dA' = \sigma_0 2\pi \rho' d\rho'$, em que o anel tem comprimento de circunferência $2\pi\rho'$, espessura $d\rho'$ e por isso $dA' = 2\pi \rho' d\rho'$ :

$$ d\vec{E} = \frac{2\pi\,\sigma_0 z_0\rho' d\rho'}{4\pi\epsilon_0 (z_0^2 + \rho'^2)^{3/2}} \hat{k} = \frac{\sigma_0 z_0\rho' d\rho'}{2\epsilon_0 (z_0^2 + \rho'^2)^{3/2}} \hat{k} $$

tal que basta integrar em relação a $\rho'$ de $0$ m ao raio $R$ do disco :

$$ \vec{E}(\vec{r}_0) = \vec{E}(z_0) = \int d\vec{E} = \int_{0\,\text{m}}^{R} \frac{\sigma_0 z_0\rho' d\rho'}{2\epsilon_0 (z_0^2 + \rho'^2)^{3/2}} \hat{k} = \frac{\sigma_0 z_0}{2\epsilon_0} \int_{0\,\text{m}}^{R} \frac{\rho' d\rho'}{(z_0^2 + \rho'^2)^{3/2}} \hat{k} $$

Essa integral escalar de uma variável é do tipo :

$$ I = \int \frac{\rho' d\rho'}{(z_0^2 + \rho'^2)^{3/2}} = - \frac{1}{\sqrt{z_0^2 + \rho'^2}} + C $$

que pode ser obtida via :

- tabela de integrais, essa é a referência mundial, vide busca na Internet por :  
[Table of Integrals, Series, and Products - I.S. Gradshteyn and I.M. Ryzhik](https://www.google.com/search?q=Gradshteyn+Ryzhik&oq=Gradshteyn+Ryzhik);

- cálculo manual usando substituição de variáveis ($u' = z_0^2 + \rho'^2$, $du' = 2\rho' d\rho'$) :

$$ I = \int \frac{du'}{2(u')^{3/2}} = \frac{1}{2} \int (u')^{-3/2} du' = \frac{1}{2 \left(\frac{-1}{2}\right)} (u')^{-1/2} + C = - \frac{1}{\sqrt{z_0^2 + \rho'^2}} + C $$

logo :

$$ \vec{E}(z_0) = \frac{\sigma_0 z_0}{2\epsilon_0} \left[ - \frac{1}{\sqrt{z_0^2 + \rho'^2}} \right]_{0\,\text{m}}^{R} \hat{k} = \frac{\sigma_0 z_0}{2\epsilon_0} \left[ - \frac{1}{\sqrt{z_0^2 + R^2}} + \frac{1}{\sqrt{z_0^2}} \right] \hat{k} $$

$$ \rightarrow \,\,\, \vec{E}(z_0) = E(z_0)\,\hat{k} = \frac{\sigma_0}{2\epsilon_0} \left[ \frac{z_0}{|z_0|} - \frac{z_0}{\sqrt{z_0^2 + R^2}} \right] \hat{k} $$

- cálculo simbólico via [softwares CAS (Computer Algebra Systems)](https://en.wikipedia.org/wiki/List_of_computer_algebra_systems), por exemplo o [módulo SymPy](https://www.sympy.org/) (em [Python](https://python.org/)) calculando a integral definida :

```python
from sympy import *
```

```python
z0, rho, sigma0, epsilon0, R = symbols('z0, rho, sigma0, epsilon0, R')
```

```python
Integral((sigma0*z0/(2*epsilon0))*rho/(z0**2 + rho**2)**(3/2), (rho, 0, R))
```

```python
integrate((sigma0*z0/(2*epsilon0))*rho/(z0**2 + rho**2)**(3/2), (rho, 0, R))
```

porém tal solução só vale para $z_0$ positivo.

#### Análise da solução (2.d) de $\vec{E}(z_0)$

Note que a função escalar $E(z_0)$ é também ímpar em relação a $z_0$, logo, para densidade superficial $\sigma_0$ positiva, $E(z_0)$ tem o mesmo sinal que $z_0$, não sendo definido em $z_0 = 0$ m, pois há uma descontinuidade em $z_0 = 0$ m :

$$ \lim_{z_0 \to 0^{+}\text{m}} \vec{E}(z_0) = \lim_{z_0 \to 0^{+}\text{m}} \frac{\sigma_0}{2\epsilon_0} \left[ \frac{z_0}{|z_0|} - \frac{z_0}{\sqrt{z_0^2 + R^2}} + \right] \hat{k} = \frac{\sigma_0}{2\epsilon_0} \left[ 1 - 0 \right] \hat{k} = \frac{\sigma_0}{2\epsilon_0} \hat{k} $$

$$ \lim_{z_0 \to 0^{-}\text{m}} \vec{E}(z_0) = \lim_{z_0 \to 0^{-}\text{m}} \frac{\sigma_0}{2\epsilon_0} \left[ \frac{z_0}{|z_0|} - \frac{z_0}{\sqrt{z_0^2 + R^2}} + \right] \hat{k} = \frac{\sigma_0}{2\epsilon_0} \left[- 1 - 0 \right] \hat{k} = - \frac{\sigma_0}{2\epsilon_0} \hat{k} $$

Esses resultados também são válidos para $R >> |z_0|$, logo um disco circular (carregado eletricamente de forma homogênea) com raio $R$ tendendo a infinito resulta em um plano infinito com $\sigma = \sigma_0$.

## Teorema das Cascas Esféricas (homogêneas) para campo elétrico $\vec{E}$

Tal teorema para campo elétrico $\vec{E}$ (pois tem também para campo gravitacional $\vec{g}$) diz que o campo elétrico $\vec{E}$ :

- interno à casca esférica é nulo;

- externo à casca esférica é igual ao campo elétrico da carga total da casca esférica posicionada no centro da casca esférica.

Em linguagem mais matemática, se a densidade superficial de cargas elétricas for constante, $\lambda = \lambda_0 = \text{cte}$, na casca esférica de raio $R$ centrada na origem, o observador (carga-teste) na posição $\vec{r}_0$ mede :

$$ \vec{E}(\vec{r}_0) = \vec{0}\,\frac{N}{C} \,\,\,\,\,\, \text{se }\,\,\, r_0 < R $$

$$ \vec{E}(\vec{r}_0) = \frac{Q\,\vec{r}_0}{4\pi\epsilon_0\,r_0^3} \,\,\,\,\,\, \text{se }\,\,\, r_0 > R $$

O campo elétrico $\vec{E}$ para uma casca esférica carregada eletricamente pode ser resolvido via integral de superfície com densidade superficial de carga elétrica, $\sigma(\vec{r}')$ na superfície $S$ da casca esférica :

$$ \vec{E}(\vec{r}_0) = \frac{1}{4\pi\epsilon_0} \oint_S \frac{\sigma(\vec{r}')\,\vec{r}\,dA'}{|\vec{r}|^3} = \frac{1}{4\pi\epsilon_0} \oint_S \frac{\sigma(\vec{r}')(\vec{r}_0-\vec{r}')\,dA'}{|\vec{r}_0-\vec{r}'|^3} $$

Ao adotar distribuição homogênea de carga elétrica na superfície $S$ da casca esférica, i. e., densidade superficial de carga elétrica, $\sigma(\vec{r}') = \sigma_0 = \text{cte}$ :

$$ \vec{E}(\vec{r}_0) = \frac{\sigma_0}{4\pi\epsilon_0} \oint_S \frac{(\vec{r}_0-\vec{r}')\,dA'}{|\vec{r}_0-\vec{r}'|^3} $$

Por exemplo, usando coordenadas esféricas $(r, \phi, \theta)$, com $0 \le \phi \le \pi$ (ângulo de inclinação em relação ao eixo $z$) e $0 \le \theta \le 2\pi$ (ângulo azimutal, no plano $xy$, a partir do eixo $x$), com o observador fora da casca esférica :

![image.png](attachment:image.png "Casca esférica")

sendo que $ dA' = h_{\phi'} h_{\theta'}\,d\theta' d\phi' = (r)(r \sin\phi') d\theta' d\phi' = r^2 \sin\phi'\,d\theta' d\phi' $, bem como $r = R$ pois a casca esférica tem raio $R$, então :

$$ \vec{E}(\vec{r}_0) = \frac{\sigma_0}{4\pi\epsilon_0} \int_0^\pi \int_0^{2\pi} \frac{(\vec{r}_0-\vec{r}')\,R^2 \sin\phi'\,d\theta'\,d\phi'}{|\vec{r}_0-\vec{r}'|^3} $$

Resta então substituir os vetores $\vec{r}_0$ e $\vec{r}$ em componentes esféricas porém usando vetores base cartesianas (tal que a diferença de vetores-posição faça sentido usando os mesmos vetores base), algebrismo e (tentativa de) resolver a integral, que não é de fácil solução.
