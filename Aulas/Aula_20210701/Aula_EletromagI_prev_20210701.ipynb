## Laplaciano

### Teoria

Laplaciano é um operador diferencial vetorial de 2a ordem, basicamente é aplicação do divergente do gradiente de uma função escalar, ou seja, atua sobre função escalar e gera função escalar.

Em coordenadas curvilíneas $u_i$ :
 
$$ lap f = \Delta f = \nabla^2 f $$

$$ \nabla^2 f = \frac{1}{h_1 h_2 h_3}\left[ \frac{\partial}{\partial u_1} \left(\frac{h_2 h_3}{h_1} \frac{\partial\,f}{\partial u_1}\right) + \frac{\partial}{\partial u_2} \left(\frac{h_1 h_3}{h_2} \frac{\partial\,f}{\partial u_2}\right) + \frac{\partial}{\partial u_3} \left(\frac{h_1 h_2}{h_3} \frac{\partial\,f}{\partial u_3}\right) \right] $$

# Visualização de Campos Vetoriais 2D e 3D via softwares

#### Wolfram Alpha

[Wolfram Alpha](https://www.wolframalpha.com/) é uma webapp (software via Internet) que permite fazer cálculos e gráficos diversos. Vários recursos são gratuitos, mas alguns pagos (Wolfram Alpha Pro).
Tem também [Wolfram Alpha Mobile](https://products.wolframalpha.com/mobile/), versão para smartphones e tablets Android, iOS, Kindle Fire e Windows Phone/Tablet. 

Para fazer gráfico 2D de campos vetoriais, basta [digitar 'plot vector field' no campo superior](https://www.wolframalpha.com/input/?i=plot+a+vector+field). Tem :
- visualização no plano (bidimensional) de campos vetoriais 1D e 2D, descritos em coordenadas cartesianas;
- opção de nome da variável cartesiana;
- opção de domínio por eixo;
- opção de cores, download da figura, etc, via Wolfram Alpha Pro. 

#### Geogebra

[Geogebra](https://www.geogebra.org/) é um conjunto de ferramentas de cálculo e visualização matemática.

Para fazer gráfico de campos vetoriais 2D, vide :
- [Vector Fields - Juan Carlos Ponce Campuzano](https://www.geogebra.org/m/QPE4PaDZ) :
    * descritos em coordenadas cartesianas;
    * opção de domínio por eixo;
    * opção de número de vetores por eixo;
    * opção de escala do vetor e da seta do mesmo;
- [Vector Fields in 2D - Juan Carlos Ponce Campuzano](https://www.geogebra.org/m/cXgNb58T) :
    * descritos em coordenadas cartesianas;
    * opção zoom (ampliação) da região, porém é centrada na origem;
    * opção de densidade de número de vetores;
    * opção de escala dos vetores;
- etc (procure via Internet).

E de campos vetoriais 3D, vide :
- [Vector fields 3D - joelm, Juan Carlos Ponce Campuzano](https://www.geogebra.org/m/qHeBsgbx) :
    * descritos em coordenadas cartesianas;
    * opção de com ou sem eixos 3D;
    * opção de escala dos vetores;
    * opção via "View" de breve animação 3D;
- etc (procure via Internet).    

#### CalcPlot3D

[CalcPlot3D](https://c3d.libretexts.org/CalcPlot3D/index.html) é uma webapp (software via Internet) com visualização tridimensional de campos vetoriais 1D, 2D e 3D, descritos em coordenadas cartesianas. Dicas :
- selecione "Vector Field" no campo "Add to graph";
- opção de número de vetores por eixo;
- opção "Restrict view to 2D" para visualizar só em um plano;
- opção de escala dos vetores;
- opção de cores para os vetores;
- etc.
